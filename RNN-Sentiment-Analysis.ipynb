# **Integer Encoding**

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data()
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
x_train = pad_sequences(x_train, padding='post', maxlen=50) # maxlen=50
x_test = pad_sequences(x_test, padding='post', maxlen=50) # maxlen=50
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(SimpleRNN(32, input_shape=(200,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
len(x_train[1])
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
from keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data()
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
x_train = pad_sequences(x_train, padding='post', maxlen=50) # maxlen=50
x_test = pad_sequences(x_test, padding='post', maxlen=50) # maxlen=50
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Embedding(10000, 2))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

hsitory = model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```
